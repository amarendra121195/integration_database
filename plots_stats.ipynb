```python
✅ Mean, Median, Mode

✅ Skewness

✅ IQR

✅ Histogram

✅ Boxplot

✅ Bar plot

✅ Pie chart

✅ Line chart

✅ Pairplot

✅ Heatmap
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew ,mode
tips = sns.load_dataset('tips')
tips
type(tips)
total_bill = tips['total_bill']
total_bill
total_bill.mean()
total_bill.median()
total_bill.mode()
skew(total_bill)
q1 = total_bill.quantile(.25)
q1
q2= total_bill.median()
q2
q3= total_bill.quantile(.75)
q3
iqr = q3-q1
iqr
lower_bound = q1-1.5*iqr
lower_bound
upper_bound = q3+1.5*iqr
upper_bound
plt.figure(figsize=(10,5))
sns.histplot(total_bill,kde=True,color="red",bins=10)
sns.boxplot(x= total_bill,color='orange')
q1
upper_bound
q2
q3
iqr
tips
sns.barplot(x = tips['day'] , y = tips['tip'],estimator=np.mean)
smoker_count = tips['smoker'].value_counts()

smoker_count
plt.pie(smoker_count,labels=smoker_count.index
        )
tips
sns.lineplot(x = tips['size'],y = tips['tip'])
sns.pairplot(tips[['total_bill', 'tip', 'size']])
corr = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
```
